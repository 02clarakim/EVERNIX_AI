```python
from datetime import date
import pandas as pd
from agent_lab.data_connectors.market_data import fetch_fundamentals
from agent_lab.agents.ackman import AckmanAgent

universe = ["AAPL","MSFT","KO","BAC","JNJ","PG"]
fund = fetch_fundamentals(universe)

ack = AckmanAgent()
today = date.today()

decisions, table = ack.decide(today, universe, fund)

rows = []
for d in decisions:
    row = fund.loc[d.symbol].to_dict()
    rows.append({
        "symbol": d.symbol,
        "Company": row.get("Company"),
        "pe_ratio": row.get("pe"),
        "roe": row.get("roe"),
        "debt_to_equity": row.get("debt_to_equity"),
        "free_cashflow": row.get("free_cashflow"),
        "sector": row.get("sector"),
        "Ackman Score": d.score,
        "Decision": d.action.name,
        "Confidence": d.confidence,
        "Justification": d.rationale
    })

ack_table = pd.DataFrame(rows).set_index("symbol")
ack_table.to_csv("ackman_decisions.csv")
print("✅ Ackman results saved to ackman_decisions.csv")
ack_table
```
